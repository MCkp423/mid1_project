```python
df = df[df['year_film'] > 2018]
```

```python
categories_to_drop = ['JEAN HERSHOLT HUMANITARIAN AWARD', 'HONORARY AWARD', 'GORDON E. SAWYER AWARD']
df = df[~df['category'].isin(categories_to_drop)]
```

```python
unique_films = df['film'].unique()
film_oscar_nominee = pd.DataFrame({'title': unique_films, 'film_oscar_nominee': 1})
```
