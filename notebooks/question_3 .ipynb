- Is there a correlation between the critical acclaim (vote_average) and commercial success (revenue) of movies?

- Assess whether highly rated movies also tend to perform well at the box office.


```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data_cleaned/merged_data.csv')
```

```python
correlation = data['vote_average'].corr(data['revenue'])

print(correlation)
#correlation coeeficient, almost no correlation
#they perform slightllyyy better
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.scatterplot(x='vote_average', y='revenue', data=data)
plt.title('Vote Average vs Revenue')
plt.yscale('log')
plt.show()
```
