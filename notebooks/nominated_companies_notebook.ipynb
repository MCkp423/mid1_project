```python
import pandas as pd
import numpy as np
```

```python
tmdb_data_url = '../data_cleaned/cleaned_data_v1.csv'
tmdb_df = pd.read_csv(tmdb_data_url)

company_nominations = {
    'A24': 17,
    'Netflix': 14,
    'Warner Bros': 12,
    'Walt Disney': 10,
    'Searchlight': 10,
    'Paramount': 9,
    'Universal': 8,
    'Focus Features': 7,
    'Neon': 4,
    'Orion Pictures/UAR': 2,
    'Sideshow': 2,
    'Sony Pictures Classics': 2
}


tmdb_df['company_nominated'] = 0

for company, nominations in company_nominations.items():
    tmdb_df.loc[tmdb_df['production_companies'].str.contains(company, na=False, case=False), 'company_nominated'] = 1
```
